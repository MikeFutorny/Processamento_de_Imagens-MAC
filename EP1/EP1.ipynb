#Função Para visualização da base de dados

```python
def print_images_by_class(metadata_path, images_dir):
    metadata_df = pd.read_csv(metadata_path)
    all_classes = set()
    for classes_str in metadata_df['Classes']: #Buscamos pelas classes diferentes existentes
        classes = classes_str.split(', ')
        all_classes.update(classes)

    class_to_print = input(f'Digite A Classe que deseja imprimir ({", ".join(all_classes)}): ')

    class_image_paths = []

    for index, row in metadata_df.iterrows():
        if class_to_print in row['Classes']:
            image_path = os.path.join(images_dir, row['File Name'])
            class_image_paths.append(image_path)

    num_images = len(class_image_paths)
    num_cols = min(num_images, 5)

    num_rows = (num_images + num_cols - 1) // num_cols

    fig, axs = plt.subplots(num_rows, num_cols, figsize=(30, 30))

    for i, image_path in enumerate(class_image_paths):
        row_idx = i // num_cols
        col_idx = i % num_cols
        image = mpimg.imread(image_path)
        axs[row_idx, col_idx].imshow(image)
        axs[row_idx, col_idx].axis('off')

    for i in range(num_images, num_rows * num_cols):
        row_idx = i // num_cols
        col_idx = i % num_cols
        axs[row_idx, col_idx].axis('off')

    plt.tight_layout()

    plt.show()

metadata_path = './content/Metadados/metadados.csv'
images_dir = './content/Dados'
print_images_by_class(metadata_path, images_dir)
```

#Tabela sumária de Dados

```python
def get_directory_size(directory):
   total_size = 0
   for dirpath, _, filenames in os.walk(directory):
       for filename in filenames:
           file_path = os.path.join(dirpath, filename)
           total_size += os.path.getsize(file_path)
   return total_size


directory_path = './content/Dados'
total_size_bytes = get_directory_size(directory_path) #Pegamos o tamanho da soma de todos os arquivos


total_size_mb = total_size_bytes / (1024 * 1024) #Transformando na unidade certa
dados = {
   'Nomes Das Classes': ['Caneta', 'Anel', 'Pingente', 'Elastico', 'Gel', 'Moeda', 'Concha', 'Fone', 'Dado', 'Pilha'],
   'Numero de Classes': '10',
   'Numero de Imagens': '72',
   'Tamanho da Base': str(total_size_mb) + " MB",
   'Resolução das imagens': "4284 Linhas por 5712 Colunas"
}


for key, value in dados.items():
   print(f'{key}: {value}')
```
